```python
# Extract the PDF files and put into Zotero
import arxiv

search = arxiv.Search(
    # Seulement les sujets IA
    query = "cat:cs.AI",
    max_results = 1,
    sort_by = arxiv.SortCriterion.SubmittedDate
)


for result in search.results():
    print("Title:", result.title)
    print("Abstract: ", result.summary)
    print("Date:", result.published)
    print("Authors: ", result.authors)
    print("Publication: ", result.journal_ref)
    print("DOI: ", result.doi)
    print("URL: ", result.pdf_url )

    
    # Create the item to add
    item = {'key': key_collection,
    'version': zot.collection(key_collection)['version']+1,
    'library': {'type': 'user',
    'id': 4986647,
    'name': 'mzavlyan'},
    'data':{'collections':[zot.collection(key_collection)],
        "title": result.title, 
        "abstractNote": result.summary, 
        "date": result.published,
        "publication":result.journal_ref,
        "doi": result.doi,
        "url": result.pdf_url}}

    zot.addto_collection(key_collection, item)
```
